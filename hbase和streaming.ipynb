```python
import happybase
from pyspark import SparkContext, SparkConf

conf = SparkConf().setMaster('local[*]').setAppName('testing')
sc = SparkContext(conf=conf)
```

```python
cnx = happybase.Connection('localhost')
```

```python
tb = cnx.table('people')
rdd1 = sc.parallelize(tb.scan())
rdd1.collect()
```

```python
rdd1.count()
```

```python
new = sc.parallelize([('id3', 'cf1:fname', 'Newman')])
for p in new.collect():
    tb.put(p[0], {p[1]: p[2]})
```

```python
from pyspark.streaming import StreamingContext
import re
ssc = StreamingContext(sc, 60)
```

```python
lines=ssc.socketTextStream('localhost', 9999)
wordcounts=lines.filter(lambda line: len(line)>0).flatMap(lambda line: re.split('\W+', line)) \
.filter(lambda word: len(word)> 0).map(lambda word: (word, 1)).reduceByKey(lambda a, b: a+b)
wordcounts.pprint()
ssc.start()
ssc.awaitTermination()
```
